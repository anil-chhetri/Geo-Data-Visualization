```python
pd.set_option('display.max_columns', 20)
summary_data = pd.merge(left=info, right=item_info, how='inner', on='Item code', left_index=False, right_index=False, indicator=True)
summary_data.head()
```

```python
#replacing space with underscore in columns Name
summary_data.columns =  ['_'.join(x.split()) for x in summary_data.columns]
```

```python
summary_data.head()
```

```python
def cat_value(row):
    val = row['Average_value']
    result = ''
    if val < row['Good(Blue)']:
        result = 'Good'
    elif val < row['Normal(Green)']:
        result = 'Normal'
    elif val < row['Bad(Yellow)']:
        result = 'Bad'
    else:
        result = 'Very Bad'
    return result
```

```python
summary_data['result'] = summary_data[['Average_value', 'Good(Blue)', 'Normal(Green)', 'Bad(Yellow)', 'Very_bad(Red)']].apply(cat_value, axis=1)
```

```python
summary_data.head()
```

```python
summary_data[summary_data['result'] == 'Very Bad']
```

```python
## selecting only columns that we need
summary = summary_data[['Station_code', 'Measurement_date', 'Item_name', 'Unit_of_measurement', 'Average_value', 'result']]
```

```python
summary.head(2)
```

```python
import numpy as np
```

```python
pivot_table = pd.pivot_table(summary, columns='Item_name', index=['Station_code', 'Measurement_date'],aggfunc=np.sum)
```

```python
pivot_table.columns = pivot_table.columns.droplevel(0)
```

```python
pivot_table.head()
```

```python
pivot_table.reset_index(inplace=True)
```

```python
pivot_table.columns.name = None
```

```python
pivot_table
```

```python
df_with_Station = pd.merge(pivot_table, stations, how='inner', left_on='Station_code', right_on='Station code')
```
